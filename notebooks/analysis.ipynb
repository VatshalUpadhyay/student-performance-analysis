# Student Performance Analysis

## Problem Statement
This project analyzes student performance data to identify key factors affecting academic scores.

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Dataset

```python
df = pd.read_csv("../data/StudentsPerformance.csv")
df.head()
```

## Data Overview
Checking basic structure and summary of dataset

```python
df.info()
df.describe()
```

## Missing Values Check

```python
df.isnull().sum()
```

## Correlation Analysis

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Analysis 1: Gender vs Scores

```python
df.groupby("gender")[["math score","reading score","writing score"]].mean()
```

## Analysis 2: Test Preparation vs Scores

```python
df.groupby("test preparation course")[["math score","reading score","writing score"]].mean()
```

## Analysis 3: Parental Education vs Math Score

```python
df.groupby("parental level of education")["math score"].mean().sort_values()
```

## Visualization

```python
sns.barplot(x="gender", y="math score", data=df)
plt.title("Math Score by Gender")
plt.savefig("../images/gender_math.png")
plt.show()
```

```python
sns.barplot(x="test preparation course", y="math score", data=df)
plt.title("Impact of Test Preparation")
plt.savefig("../images/test_prep.png")
plt.show()
```

```python
sns.violinplot(x="gender", y="math score", data=df)
plt.title("Distribution of Math Scores by Gender")
plt.savefig("../images/math_distribution_gender.png") 

plt.show()
```

```python
sns.barplot(x="parental level of education", y="math score", data=df)
plt.xticks(rotation=45)
plt.title("Parental Education vs Math Score")
plt.savefig("../images/parent_education.png")
plt.show()
```

## Key Insights

- Students who completed test preparation scored approximately 15–20% higher  
- Female students performed better in reading and writing  
- Parental education positively impacts performance  
- Math scores correlate strongly with reading and writing
